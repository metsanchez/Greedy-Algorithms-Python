# Prim Algorithm

```python
"""
*** PRIM's Algoritması *** 

* Cycle oluşturmayacak 
* En kısa yol seçilmeli

1. Başlangıç düğümünü seç
2. Yakın komşularına bak en kısa yol olana ve daha önce gidilmemiş olana git.
3. Artık yeni birleşim yerlerini bir kapsam olarak düşünüp 2. adıma dön. Eğer bütün düğümler ziyaret edilmişse bitir. 
"""

INF = 999999

N = 5

# Komşuluk Matrisi 
G = [[0, 19, 5, 0, 0],
     [19, 0, 5, 9, 2],
     [5, 5, 0, 1, 6],
     [0, 9, 1, 0, 1],
     [0, 2, 6, 1, 0]]

selected_node = [0,0,0,0,0]
counter = 0 

# ilk düğüm yani başlangıç düğümü için ziyaret edildi yapılır. 
selected_node[0] = True
print("*** PRIM's Algoritması ***")

# Şimdi de aşağıdaki kod parçasında her bir düğüm için diğer düğümlere bakılıp min uzaklık ve ziyaret edilme durumuna göre 
# bir seçip yapılacaktır.
while counter < N -1:
    minimum = INF
    a = 0
    b = 0
    for m in range(N):
        # Ziyaret edilmiş bir düğüm için diğer düğümlerle olan uzaklıklarına bakıyoruz.
        if selected_node[m]:
            # Bu kısımda her bir düğüm için en kısa ve ziyaret edilmemiş düğümü bulmaya çalışıyoruz. Yani her düğüm için
            # yaptığımızdan dolayı bir kapsam olarak görüyoruz. 
            for n in range(N):
                # Eğer node seçilmediyse ve komşuluk matrisinde uzunluğu 0 değilse (0 da False, diğer sayılarda True döner)
                if ((not selected_node[n]) and G[m][n]):
                    if G[m][n] < minimum:
                        minimum = G[m][n]
                        a = m
                        b = n
    print(f"Birinci Düğüm:{a} - ikinci Düğüm:{b}, Arasındaki en kısa uzaklık:{minimum}")
    # Yeni gidilen düğümü ziyaret edilmiş olarak işaretledik. Birinci düğüm zaten işaretleniyor ve kodda ona göre seçiliyor.
    selected_node[b] = True
    counter +=1
                    
```

# Kruskal's Algorithm

```python
"""
*** Kruskal ***

1. Baştan sona en her kenar ayrı ayrı yazılır.
2. Kenarlara göre küçükten büyüğe sıralanır. 
3. Ek kısa kenardan uzuna doğru edge'ler eklenmeye başlar.
4. Eğer düğümler bir cycle oluşturuyorsa op edge kabul edilmez devam edilir.
"""

class Graph:
 
    def __init__(self, vertices):
        self.V = vertices  # No. of vertices
        self.graph = []
        # to store graph
 
    # function to add an edge to graph
    def addEdge(self, u, v, w):
        self.graph.append([u, v, w])
 
    # Parent'ını çağır
    def find(self, parent, i):
        if parent[i] != i:
            parent[i] = self.find(parent, parent[i])
        return parent[i]
 

    def union(self, parent, rank, x, y):

        # rank değerine göre parentlarını değiştiriyoruz.
        if rank[x] < rank[y]:
            parent[x] = y
        elif rank[x] > rank[y]:
            parent[y] = x
 
        else:
            parent[y] = x
            rank[x] += 1
 

    def KruskalMST(self):
 
        result = []  
        i = 0
        e = 0
        # Ağırlıkları küçükten büyüğe sıralıyoruz
        self.graph = sorted(self.graph,
                            key=lambda item: item[2])
 
        parent = []
        rank = []
 
        # Create V subsets with single elements
        for node in range(self.V):
            parent.append(node)
            rank.append(0)
 
        # Number of edges to be taken is less than to V-1
        while e < self.V - 1:
 
            # graph'ı küçükten büyüğe sıraladığımız için 1. yanı en kısa olan edge'i seçiyoruz ve i değerini bir arttırıyoruz.
            u, v, w = self.graph[i]
            i = i + 1
            x = self.find(parent, u)
            y = self.find(parent, v)
 
            # Cycle oluşturmaması için bir kontrol yapıyoruz
            if x != y:
                e = e + 1
                result.append([u, v, w])
                self.union(parent, rank, x, y)
 
        minimumCost = 0
        print("Edges in the constructed MST")
        for u, v, weight in result:
            minimumCost += weight
            print("%d -- %d == %d" % (u, v, weight))
        print("Minimum Spanning Tree", minimumCost)
 
 
# Driver's code
if __name__ == '__main__':
    g = Graph(4)
    g.addEdge(0, 1, 10)
    g.addEdge(0, 2, 6)
    g.addEdge(0, 3, 5)
    g.addEdge(1, 3, 15)
    g.addEdge(2, 3, 4)
 
    # Function call
    g.KruskalMST()
```

# Dijkstra’s Algorithm

```python
class Graph():
 
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for column in range(vertices)]
                      for row in range(vertices)]
 

    def minDistance(self, dist, sptSet):
 
        # Min a sonsuz bir sayı koyulur
        min = 1e7
 
        # En küçük uzaklığı bulma 
        for v in range(self.V):
            if dist[v] < min and sptSet[v] == False:
                min = dist[v]
                min_index = v
 
        return min_index
 
    # Function that implements Dijkstra's single source
    # shortest path algorithm for a graph represented
    # using adjacency matrix representation
    def dijkstra(self, src):
        
        # Uzaklıkları ilk başta hepsi için sonsuz yapıyoruz 1e7 = 10^7
        dist = [1e7] * self.V
        # Başlangıç düğümümüzün uzaklığını 0 yapıyoruz
        dist[src] = 0
        # Bütün ziyaret edilme durumları false yapılır.
        sptSet = [False] * self.V
 
        for count in range(self.V):
 
            u = self.minDistance(dist, sptSet)
            
            # Seçilen düğüm ziyaret edilmiş olarak işaretlenir. 
            sptSet[u] = True
 
            # Eğer seçilen değer küçükşe durumu
            for v in range(self.V):
                if (self.graph[u][v] > 0 and
                   sptSet[v] == False and
                   dist[v] > dist[u] + self.graph[u][v]):
                    dist[v] = dist[u] + self.graph[u][v]
 
        print("Vertex \t Distance from Source")
        for node in range(self.V):
            print(node, "\t\t", dist[node])

    
g = Graph(9)
g.graph = [[0, 4, 0, 0, 0, 0, 0, 8, 0],
           [4, 0, 8, 0, 0, 0, 0, 11, 0],
           [0, 8, 0, 7, 0, 4, 0, 0, 2],
           [0, 0, 7, 0, 9, 14, 0, 0, 0],
           [0, 0, 0, 9, 0, 10, 0, 0, 0],
           [0, 0, 4, 14, 10, 0, 2, 0, 0],
           [0, 0, 0, 0, 0, 2, 0, 1, 6],
           [8, 11, 0, 0, 0, 0, 1, 0, 7],
           [0, 0, 2, 0, 0, 0, 6, 7, 0]]
 
g.dijkstra(0)
```

# Boruvka's Algorithm

```python
from collections import defaultdict
 
class Graph:
 
    def __init__(self,vertices):
        self.V= vertices 
        self.graph = [] 
  
    def addEdge(self,u,v,w):
        self.graph.append([u,v,w])
 
    def find(self, parent, i):
        if parent[i] != i:
            self.find(parent, parent[i])
        return parent[i]
 
    def union(self, parent, rank, x, y):
        xroot = self.find(parent, x)
        yroot = self.find(parent, y)
 
        # burada birleştirme işlemi yapıyoruz. Eğer rank olarak büyük olanın parentını az olan da alır
        if rank[xroot] < rank[yroot]:
            parent[xroot] = yroot
        elif rank[xroot] > rank[yroot]:
            parent[yroot] = xroot
 
        else :
            parent[yroot] = xroot
            rank[xroot] += 1
 
    # The main function to construct MST using Kruskal's algorithm
    def boruvkaMST(self):
        parent = []; rank = [];
 
        cheapest =[]
 
        numTrees = self.V
        MSTweight = 0
 
        for node in range(self.V):
            parent.append(node)
            rank.append(0)
            cheapest =[-1] * self.V
     
 
        while numTrees > 1:
 
            # Traverse through all edges and update
               # cheapest of every component
            for i in range(len(self.graph)):
 
                # Find components (or sets) of two corners
                # of current edge
                u,v,w =  self.graph[i]
                set1 = self.find(parent, u)
                set2 = self.find(parent ,v)
 
                if set1 != set2:    
                     
                    if cheapest[set1] == -1 or cheapest[set1][2] > w :
                        cheapest[set1] = [u,v,w]
 
                    if cheapest[set2] == -1 or cheapest[set2][2] > w :
                        cheapest[set2] = [u,v,w]
 
            for node in range(self.V):
 
                #Check if cheapest for current set exists
                if cheapest[node] != -1:
                    u,v,w = cheapest[node]
                    set1 = self.find(parent, u)
                    set2 = self.find(parent ,v)
 
                    if set1 != set2 :
                        MSTweight += w
                        self.union(parent, rank, set1, set2)
                        print ("Edge %d-%d with weight %d included in MST" % (u,v,w))
                        numTrees = numTrees - 1
             
            #reset cheapest array
            cheapest =[-1] * self.V
 
             
        print ("Weight of MST is %d" % MSTweight)
                           
 
     
g = Graph(4)
g.addEdge(0, 1, 10)
g.addEdge(0, 2, 6)
g.addEdge(0, 3, 5)
g.addEdge(1, 3, 15)
g.addEdge(2, 3, 4)
 
g.boruvkaMST()
```
